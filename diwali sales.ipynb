```python
import matplotlib.pyplot as plt

# Top 10 most sold products
fig1, ax1 = plt.subplots(figsize=(12, 7))

# Group by Product_ID, sum Orders, and plot top 10
df.groupby('Product_ID')['Orders'].sum().nlargest(10).sort_values(ascending=False).plot(kind='bar', color='skyblue', ax=ax1)

# Add labels to the bars
for p in ax1.patches:
    ax1.annotate(f'{p.get_height():,.0f}', (p.get_x() + p.get_width() / 2., p.get_height()), 
                 ha='center', va='center', fontsize=12, color='black', xytext=(0, 5), textcoords='offset points')

# Set title and labels
ax1.set_title('Top 10 Most Sold Products', fontsize=16, fontweight='bold')
ax1.set_xlabel('Product ID', fontsize=14)
ax1.set_ylabel('Number of Orders', fontsize=14)

# Rotate x-axis labels for better readability
plt.xticks(rotation=45, ha='right', fontsize=12)

# Show the plot
plt.tight_layout()
plt.show()
```
